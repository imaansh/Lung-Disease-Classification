Reading Gamma Corrected Images

```python
data_dir = '/content/drive/MyDrive/Notebooks/Training 1'
train_ds = tf.keras.utils.image_dataset_from_directory(
  data_dir,
  labels='inferred',
  label_mode='int',
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
val_dir = '/content/drive/MyDrive/Notebooks/Validation1'
val_ds = tf.keras.utils.image_dataset_from_directory(
  val_dir,
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)


```

Splitting Validation data into val_data and test data


```python
def split_data(ds, test_split = 0.2, val_split=0.8, shuffle_size=10000):
    ds = ds.shuffle(shuffle_size, seed=123)
    total_length = len(ds)
    test_ds = ds.take(int(total_length*test_split))
    validation_ds = ds.skip(int(total_length*test_split)).take(int(total_length * val_split))

    return test_ds,validation_ds
```

```python
test_d, val_d = split_data(val_ds)
```

Class Names

```python
class_names = (train_ds.class_names)
print(class_names)
```

Data Lengths

```python
print(len(train_ds))
print(len(val_d))
print(len(test_d))
```

Visualising Data


```python
plt.figure(figsize=(10,10))


for image, label in train_ds.take(2):
    for i in range(2):
        plt.subplot(1,2, i+1)
        plt.imshow(image[i].numpy().astype("uint8"))
        plt.title(class_names[label[i]])
        plt.axis('off')

```

Normalizing Data

```python
resize_and_rescale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(img_height,img_width),
    layers.experimental.preprocessing.Rescaling(scale= 1./255)

])
```

```python
data_augmentation = tf.keras.Sequential([
  layers.RandomFlip("horizontal_and_vertical"),
])
```

Model Training

```python
num_classes = 6
shape = (batch_size,img_height,img_width,3)
model = tf.keras.Sequential([
        resize_and_rescale,

        layers.Conv2D(filters= 64, kernel_size=(3,3),padding='same', activation='relu',input_shape = shape),
        layers.MaxPooling2D((2,2)),

        layers.Conv2D(filters= 64, kernel_size=(3,3),padding='same', activation='relu'),
        layers.MaxPooling2D((2,2)),

        layers.Conv2D(filters= 128, kernel_size=(3,3),padding='same', activation='relu'),
        layers.MaxPooling2D((2,2)),

        layers.Conv2D(filters= 128, kernel_size=(3,3),padding='same', activation='relu'),
        layers.MaxPooling2D((2,2)),

        layers.Flatten(),
        layers.Dropout(0.2),
        layers.Dense(128, activation='relu',kernel_regularizer=regularizers.l2(0.01)),
        layers.Dense(num_classes, activation='softmax',kernel_regularizer=regularizers.l2(0.01)),

])
```

```python
model.compile(
  optimizer='adam',
  loss=tf.losses.SparseCategoricalCrossentropy(from_logits=False),
  metrics=['accuracy'])
```

```python
history = model.fit(
  train_ds,
  validation_data=val_d,
  epochs=30
)
```

Model Evaluation

```python
output = model.evaluate(test_d)
```
